```python
plt.figure(figsize=(12, 6))
appData.plot(  y='user_age',   x='rating',   kind= 'bar' )
# sns.scatterplot(data=appData, x='user_age', y='rating', alpha=0.5)
plt.title('Relationship between User Age and Rating')
plt.xlabel('User Age')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=appData, x='device_type', order=appData['device_type'].value_counts().index)
plt.title('Distribution of Reviews by Device Type')
plt.xlabel('Device Type (1=Android Tablet, 2=iPad, 3=iOS, 4=Windows Phone, 5=Android)')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=appData, x='device_type', y='rating', errorbar=None)
plt.title('Average Rating by Device Type')
plt.xlabel('Device Type (1=Android Tablet, 2=iPad, 3=iOS, 4=Windows Phone, 5=Android)')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
userCountryCount.plot(kind='bar')
plt.title('Distribution of Reviews by Country')
plt.xlabel('Country')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
average_rating_by_country = appData.groupby('user_country')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
average_rating_by_country.plot(kind='bar')
plt.title('Average Rating by Country')
plt.xlabel('Country')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
